```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
data=pd.read_csv('Sales_data_zone_wise.csv')
data
```

```python
#The mean sales generated by each zone.
data.mean(numeric_only=True).apply('{:.5f}'.format)
```

```python
#Total sales generated by all the zones for each month.
```

```python
data_1=data.groupby('Month').sum()
data_1['Total Sales']= data_1.sum(axis=1)
data_1
```

```python
#Check whether all the zones generate the same amount of sales.
```

```python
# Since we want to check the amount of sales for all zones we will use One Way Anova
# Hypothesis
#H0:All zones generate same amount of sales
#H1:All zones generate different amount of sales
```

```python
fvalue,pvalue = stats.f_oneway(data_1['Zone - A'],data_1['Zone - B'],data_1['Zone - C'],data_1['Zone - D'])
print(fvalue,pvalue)
```

```python
if pvalue<0.05:
   print('Reject the null hypothesis')
else:
   print('Fail to reject the null hypothesis')
```

#Since p value is less than .05 we reject the null hypothesis and we came to the conclusion that the amount of 
sales for all zones are different
